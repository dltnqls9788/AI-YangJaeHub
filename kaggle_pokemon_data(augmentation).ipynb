## Augmentation

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator 

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.1
)
```

```python
train_generator = train_datagen.flow_from_dataframe(
    new_df,
    x_col='filepath', y_col='Type',
    target_size=(120,120),
    color_mode='rgba', # r g b + alpha, opacity 불투명 
    class_mode='categorical',
    batch_size=32

)
```

```python
Image_sample = train_generator.next()[0]

plt.figure(figsize=(10,10))

for i in range(9):
  plt.subplot(3,3, i+1)
  plt.imshow(Image_sample[i, :,:,:])
  plt.axis('off')
```

```python
validation_generator = train_datagen.flow_from_dataframe(
    new_df,
    x_col='filepath', y_col='Type',
    target_size=(120,120),
    color_mode='rgba', # r g b + alpha, opacity 불투명 
    class_mode='categorical',
    batch_size=4,
    shuffle=True, 
    seed=1,
    subset='validation'

)
```

## Model 

```python
from tensorflow import keras as ks 

model = ks.models.Sequential()

model.add(ks.layers.Dense(4, input_shape=(120,120,4)))

model.add(ks.layers.Conv2D(64, (7,7), activation='relu'))
model.add(ks.layers.MaxPool2D())

model.add(ks.layers.Conv2D(128, (7,7), activation='relu'))
model.add(ks.layers.MaxPool2D())

model.add(ks.layers.Conv2D(256, (7,7), activation='relu'))
model.add(ks.layers.MaxPool2D())

model.add(ks.layers.Conv2D(512, (7,7), activation='relu'))
model.add(ks.layers.MaxPool2D())

model.add(ks.layers.Flatten())
model.add(ks.layers.Dense(1024,activation='relu'))
model.add(ks.layers.Dropout(0.5))
model.add(ks.layers.Dense(3, activation='softmax'))


model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# Callback to stop training 일정 정확도가 도달했을 때 
class myCallback(ks.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('acc')>0.7) and (logs.get('val_acc')>0.7):
      print('\n reached 70% accuracy so stopping training')
      self.model.stop_training = True

callbacks = myCallback()
```

```python
history = model.fit(
    train_generator,
    validation_data = validation_generator,
    batch_size = 20,
    epochs=50
    #callbacks=[ks.callbacks.ReduceLROnPlateau(),
    #           callbacks]
)
```

```python
plt.style.use('ggplot')

acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.figure(figsize=(6, 5))

plt.plot(epochs, acc, 'r', label='training_accuracy')
plt.plot(epochs, val_acc, 'b', label='validation_accuracy')
plt.title('Training and Validation Accuarcy')
plt.xlabel('-----epochs--->')
plt.ylabel('Accuracy')
plt.legend()

plt.figure(figsize=(6, 5))

plt.plot(epochs, loss, 'r', label='training_loss')
plt.plot(epochs, val_loss, 'b', label='validation_loss')
plt.title('Training and Validation Loss')
plt.xlabel('----epochs--->')
plt.ylabel('Loss')
plt.legend()

plt.show()
```
